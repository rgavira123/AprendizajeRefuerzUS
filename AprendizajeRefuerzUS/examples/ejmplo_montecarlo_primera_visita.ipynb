```python
import AprendizajeRefuerzUS
```

```python
from AprendizajeRefuerzUS import problem as prob

import pkg_resources

mapa1 = pkg_resources.resource_filename('AprendizajeRefuerzUS', 'maps/map1.txt')
mapa2 = pkg_resources.resource_filename('AprendizajeRefuerzUS', 'maps/map2.txt')
mapa3 = pkg_resources.resource_filename('AprendizajeRefuerzUS', 'maps/map3.txt')

problem = prob.Problem(mapa1,0.2)
```

```python
transiciones = problem.transiciones
recompensas = problem.recompensas
```

```python
from AprendizajeRefuerzUS.algorithms import util

politica_greedy = problem.politica

politica_procesable = util.obtener_politica_final(politica_greedy)
```

```python
from AprendizajeRefuerzUS.algorithms import Montecarlo as mc

mc = mc.MonteCarlo(transiciones, recompensas, politica0= politica_procesable , factor_descuento= 0.9, max_iteraciones=2000)

mc.entrenar_primera_visita()
```

```python
politica_mc = mc.obtener_politica()
problem.actualiza_politica(politica_mc)
```
